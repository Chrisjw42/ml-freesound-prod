```python
def get_result(folder, fn, model):
    sample_df = pd.read_csv("../data/freesound-ml/{}/{}".format(folder, fn))
    
    # Drop dodgy index row
    sample_df.drop(sample_df.columns[0], axis=1, inplace=True)
    
    y = sample_df["class"]
    testX = sample_df.drop("class", axis=1)
    pred = mode(model.predict_classes(testX))[0][0]
    
    row = {}
    row["pred"] = pred
    row["y"] = y.iloc[0]
    row["correct"] = int(pred == y.iloc[0])
    result = pd.DataFrame(data=row, index=[fn])
    return result
```
